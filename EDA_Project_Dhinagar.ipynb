üì¶ Importing Libraries


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style='whitegrid')
```

üìÇ Loading and Exploring Data


```python


# Replace with your file name, e.g., 'bank.csv' or 'bank-full.csv'
df = pd.read_csv("bank-full.csv", sep=';')

# Basic info
print("Shape of dataset:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nSummary Stats:\n", df.describe(include='all'))
```

```python
# üîç Missing Values & Duplicates
# ----------------------------------------
print("\nMissing values:\n", df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())

# Remove duplicates if any
df = df.drop_duplicates()
```

üìä Visualizations


```python


# Histograms for numeric features
df.hist(bins=20, figsize=(15, 10), color='skyblue')
plt.suptitle("Distributions of Numeric Features")
plt.tight_layout()
plt.show()


```

```python
# Bar plot for categorical features
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, palette='pastel')
    plt.title(f"Count of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

üì¶ Group and Feature Analysis


```python

# Group by job and get average balance
grouped = df.groupby('job')['balance'].mean().sort_values(ascending=False)
print("\nAverage Balance by Job:\n", grouped)

# Visualizing correlation
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Pairplot of key features
sns.pairplot(df[['age', 'balance', 'duration', 'campaign', 'pdays']], corner=True)
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
# üõ† Feature Engineering
# ----------------------------------------

# Convert 'month' and 'day_of_week' to ordered categories if needed
# (Note: this depends on the version you're using)

# Example: Categorize 'age' into bins
df['age_group'] = pd.cut(df['age'], bins=[0, 25, 40, 60, 100],
                         labels=['Youth', 'Young Adult', 'Adult', 'Senior'])

# Duration in minutes
df['duration_mins'] = df['duration'] / 60

# Check new features
df[['age', 'age_group', 'duration', 'duration_mins']].head()
```

```python
 Advanced Python Techniques
```

```python


# Lambda function: Flag high balance
df['high_balance'] = df['balance'].apply(lambda x: 'High' if x > 1000 else 'Low/Medium')

# UDF: Label campaign effectiveness
def campaign_score(campaign):
    if campaign <= 1:
        return 'Effective'
    elif campaign <= 3:
        return 'Moderate'
    else:
        return 'Ineffective'

df['campaign_effectiveness'] = df['campaign'].apply(campaign_score)

# List comprehension: Get columns with missing values
missing_cols = [col for col in df.columns if df[col].isnull().sum() > 0]
print("Columns with missing values:", missing_cols)
```


üíæ Saving Cleaned Dataset
